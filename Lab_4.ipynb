## Using learnt representations
In practice, very few people train an entire Convolutional Network from scratch (with random initialization), because it is relatively rare to have a dataset of sufficient size. Instead, it is common to pretrain a ConvNet on a very large dataset (e.g. ImageNet, which contains 1.2 million images with 1000 categories), and then use the ConvNet either as an initialization or a fixed feature extractor for the task of interest.


We'll train a model to classify ants and bees. We have about 120 training images each for ants and bees. There are 75 validation images for each class. Usually, this is a very small dataset to generalize upon, if trained from scratch. Since we are using transfer learning, we should be able to generalize reasonably well.

This dataset is a very small subset of imagenet.




```python
# Import packages
import torch
import torchvision
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, SubsetRandomSampler
import torch.nn as nn
from torch import optim
import torch.nn.functional as F

import matplotlib.pyplot as plt
import numpy as np
import time
import os
```

```python
# Device configuration (whether to run on GPU or CPU)
device = 'cpu' #torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Download and extract dataset
!wget https://download.pytorch.org/tutorial/hymenoptera_data.zip
!unzip -q hymenoptera_data.zip
```

```python
# Data augmentation and normalization for training
# Just normalization for validation
train_transform = transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
val_transform = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

data_dir = './hymenoptera_data'
train_dataset = ImageFolder(os.path.join(data_dir, 'train'), train_transform)
val_dataset = ImageFolder(os.path.join(data_dir, 'val'), val_transform)
train_dataloader = DataLoader(train_dataset, batch_size=4,shuffle=True, num_workers=2)
val_dataloader = DataLoader(val_dataset, batch_size=4,shuffle=True, num_workers=2)
class_names = train_dataset.classes
```

```python
def imshow(inp, title=None):
    """Imshow for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)  # pause a bit so that plots are updated


# Get a batch of training data
inputs, classes = next(iter(train_dataloader))

# Make a grid from batch
out = torchvision.utils.make_grid(inputs)

imshow(out, title=[class_names[x] for x in classes])
```

```python
def train_model(model, train_dataloader, criterion, optimizer, num_epochs=25):

    for epoch in range(num_epochs):
        print(f'Epoch {epoch}/{num_epochs - 1}')
        print('-' * 10)
        model.train()
        running_loss = 0.0
        running_corrects = 0

        # Iterate over data.
        for i, (inputs, labels) in enumerate(train_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            # zero the parameter gradients
            optimizer.zero_grad()

            # forward
            # track history if only in train
            with torch.set_grad_enabled(True):
              outputs = model(inputs)
              _, preds = torch.max(outputs, 1)
              loss = criterion(outputs, labels)

              # backward + optimize only if in training phase
              loss.backward()
              optimizer.step()

            # statistics
            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)

        epoch_loss = running_loss / len(train_dataloader.dataset)
        epoch_acc = running_corrects.double() / len(train_dataloader.dataset)

        print(f'Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

        print()
```

```python
def visualize_model(model, num_images=10):
    was_training = model.training
    model.eval()
    images_so_far = 0
    fig = plt.figure()

    with torch.no_grad():
        for i, (inputs, labels) in enumerate(val_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            for j in range(inputs.size()[0]):
                images_so_far += 1
                ax = plt.subplot(num_images//2, 2, images_so_far)
                ax.axis('off')
                ax.set_title(f'predicted: {class_names[preds[j]]}')
                imshow(inputs.cpu().data[j])

                if images_so_far == num_images:
                    model.train(mode=was_training)
                    return
        model.train(mode=was_training)
```

```python
from torchvision import models

model_ft = models.resnet18(pretrained=True)
num_ftrs = model_ft.fc.in_features
# Here the size of each output sample is set to 2.
# Alternatively, it can be generalized to nn.Linear(num_ftrs, len(class_names)).
model_ft.fc = nn.Linear(num_ftrs, 2)

model_ft = model_ft.to(device)

loss_func = nn.CrossEntropyLoss()

# Observe that all parameters are being optimized
optimizer = optim.SGD(model_ft.parameters(), lr=0.001, momentum=0.9)
```

```python
num_epochs = 5
train_model(model_ft, train_dataloader, loss_func, optimizer, num_epochs)
```

```python
visualize_model(model_ft)
```

1) What is the significance of using data augmentations like resize, crop, etc. on training data?
Data augmentation is a critical technique in computer vision, especially when working with limited training data. Models like CNNs are highly data-hungry—they perform best when exposed to a large number of varied examples. When that's not possible, augmentation helps bridge the gap.

By applying transformations like random resizing, cropping, horizontal flipping, rotation, or even color jittering, we simulate different scenarios in which an object might appear in the real world. For instance, a bee might not always be in the center of the image or might be flipped, partially visible, or in varying lighting conditions. These augmentations help the model become invariant to such changes and improve its ability to generalize.

Furthermore, augmentation acts like a regularizer—it introduces enough randomness during training to prevent the model from overfitting to the exact pixel-level details of the training images. Essentially, you're exposing the model to "new" data every epoch, helping it learn more meaningful and flexible representations.

2) What performance do you get if you don't use a pretrained ResNet model?
Training a ResNet from scratch (i.e., with pretrained=False) on a small dataset like ants vs bees is usually not ideal. ResNet is a deep network with millions of parameters. Without sufficient data, it's extremely hard to train effectively from scratch because:

The model doesn't know any useful features to start with—it begins with random weights.

The training data may not be enough for it to discover useful patterns on its own.

As a result, it learns slowly, is more likely to overfit, and generally performs worse in terms of accuracy.

In practice, you might see accuracies drop significantly, sometimes below 65–70%, depending on the training setup and number of epochs. In contrast, a pretrained ResNet model (trained on ImageNet) already knows how to detect general visual features, which can be reused and fine-tuned for your task. This usually results in much faster convergence and higher accuracy—often in the 90%+ range for the ants vs bees classification problem.

So, pretrained models are especially valuable when you're working with small, domain-specific datasets and want good performance without needing massive compute or data.

3) If the ResNet model was pretrained on a dataset significantly different than the ants vs bees data, would you still get good performance?
This dives into the concept of transfer learning. The effectiveness of transfer learning depends largely on the similarity between the source domain (the data the model was pretrained on) and the target domain (your dataset).

ResNet models pretrained on ImageNet typically work well across many vision tasks, because ImageNet is broad—it contains animals, objects, textures, etc. But if ResNet were pretrained on a very different kind of dataset (like medical X-rays, satellite imagery, or handwriting), then the learned features may not transfer as cleanly.

Still, there’s a catch: the early layers of CNNs generally learn very generic features—like edges, blobs, textures, and basic shapes—which are common to many types of images. These can still be useful, even if the higher layers (which learn more specific features) are less transferable.

In that case, a common strategy is to:

Freeze the early layers (keep them unchanged),

Replace and retrain the deeper layers,

Optionally fine-tune the whole network using a lower learning rate.

This way, you can still take advantage of the general features learned from the source dataset, even if the domains aren’t closely related. You might not get top-tier performance right away, but it often still beats training from scratch.
